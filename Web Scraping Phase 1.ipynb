```python
import re
import regex as re
import selenium
import pandas as pd
import numpy as np
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import requests
import time

BHK=[]
Property=[]
Beds=[]
Baths=[]
Balcony=[]
Address=[]
City=[]
State=[]
Country=[]
Landmarks=[]
Carpet=[]
Status=[]
Floor=[]
Transaction=[]
Year=[]
Price=[]
Furnished=[]
Facing=[]
Parking=[]
Ownership=[]
Booking=[]
Buy=[]
```

```python
Cities=['Chicago','Seattle','Los Angeles','San Diego','Portland','Washington','San Francisco','New York','Austin','Las Vegas']

#1 Scrap data from https://www.christiesrealestate.com/
property_urls11=[]
for loc in Cities:
    driver=webdriver.Chrome()
    driver.get("https://www.christiesrealestate.com/sales/int")
    time.sleep(3)
    location=driver.find_element(By.XPATH,'//*[@id="skiptomaincontentmoduletarget"]')
    location.send_keys(loc)
    time.sleep(3)
    try:
        search=driver.find_element(By.XPATH,'//div[@class="category-container category-undefined grid__item"]/dl[1]/dd')
        search.click()
        for _ in range(50):
            driver.execute_script("window.scrollBy(0,20)")
            time.sleep(3)
        start=0
        end=1
        for page in range(start,end):
            url11=driver.find_elements(By.XPATH,'//div[@class="listing-item__text--bg  listing-item__text--flex  listing-item__details--border  js-same-height"]/a')
            for i in url11:
                property_urls11.append(i.get_attribute("href"))
            time.sleep(3)
    except NoSuchElementException:
        driver.close()

print(property_urls11)
for url11 in property_urls11:
    driver=webdriver.Chrome()
    driver.get(url11)
    time.sleep(5)
        
    try:
        bhk=driver.find_element(By.XPATH,'//*[@id="listinginfo"]/div/div/div/div[1]/div[2]/div[1]/div/dl/dd')
        BHK.append(''.join(i for i in bhk.text if i.isdigit()))
    except NoSuchElementException:
        BHK.append(np.nan)
            
    try:
        property=driver.find_element(By.XPATH,'//*[@id="listinginfo"]/div/div/div/div[3]/div[2]/div[1]/div/dl/dd')
        Property.append(property.text)
    except NoSuchElementException:
        Property.append(np.nan)
            
    try:
        beds=driver.find_element(By.XPATH,'//*[@id="listinginfo"]/div/div/div/div[1]/div[2]/div[1]/div/dl/dd')
        Beds.append(''.join(i for i in beds.text if i.isdigit()))
    except NoSuchElementException:
        Beds.append(np.nan)
            
    try:
        baths=driver.find_element(By.XPATH,'//*[@id="listinginfo"]/div/div/div/div[1]/div[2]/div[2]/div/dl/dd')
        Baths.append(''.join(i for i in baths.text if i.isdigit()))
    except NoSuchElementException:
        Baths.append(np.nan)
    
    Balcony.append(np.nan)
    
    try:
        address=driver.find_element(By.XPATH,'//*[@id="listingtitle"]/h1/div[1]')
        Address.append(address.text)
    except NoSuchElementException:
        Address.append(np.nan)
        
    try:
        city=driver.find_element(By.XPATH,'//*[@id="listingtitle"]/h1/div[2]/div/span[3]')
        City.append(city.text)
    except NoSuchElementException:
        City.append(np.nan)
        
    try:
        state=driver.find_element(By.XPATH,'//*[@id="listingtitle"]/h1/div[2]/div/span[5]')
        State.append(state.text)
    except NoSuchElementException:
        State.append(np.nan)
    
    Country.append('USA')
    
    try:
        landmarks=driver.find_element(By.XPATH,'//*[@id="listingtitle"]/h1/div[2]/div/span[1]')
        Landmarks.append(landmarks.text)
    except NoSuchElementException:
        Landmarks.append(np.nan)
    
    try:
        carpet1=driver.find_element(By.XPATH,'//*[@id="listinginfo"]/div/div/div/div[1]/div[2]/div[3]/div/dl/dt')
        if carpet1.text=="Interior":
            carpet=driver.find_element(By.XPATH,'//*[@id="listinginfo"]/div/div/div/div[1]/div[2]/div[3]/div/dl/dd')
            Carpet.append(''.join(i for i in carpet.text if i.isdigit()))
        else:
            carpet=driver.find_element(By.XPATH,'//*[@id="listinginfo"]/div/div/div/div[1]/div[2]/div[4]/div/dl/dd')
            Carpet.append(''.join(i for i in carpet.text if i.isdigit()))
    except NoSuchElementException:
        Carpet.append(np.nan)
        
    Status.append(np.nan)
        
    Floor.append(np.nan)
        
    try:
        transaction=driver.find_element(By.XPATH,'//*[@id="listingtitle"]/div[2]/div')
        Transaction.append(transaction.text)
    except NoSuchElementException:
        Transaction.append(np.nan)
        
    try:
        year=driver.find_element(By.XPATH,'//*[@id="listinginfo"]/div/div/div/div[3]/div[2]/div[2]/div/dl/dd')
        Year.append(''.join(i for i in year.text if i.isdigit()))
    except NoSuchElementException:
        Year.append(np.nan)
        
    try:
        price=driver.find_element(By.XPATH,'//*[@id="listinginfo"]/div/div/div/div[2]/div[2]/div[2]/div/dl/dd/span')
        Price.append(''.join(i for i in price.text if i.isdigit()))
    except NoSuchElementException:
        Price.append(np.nan)
        
    Furnished.append(np.nan)
        
    Facing.append(np.nan)
        
    Parking.append(np.nan)
        
    Ownership.append(np.nan)
        
    try:
        buy=driver.find_element(By.XPATH,'//*[@id="listingtitle"]/div[1]/span')
        Buy.append(''.join(i for i in buy.text if i.isdigit()))
        Booking.append(np.nan)
    except NoSuchElementException:
        Buy.append(np.nan)
        Booking.append(np.nan)
```

```python
Cities=['Chicago','Seattle','Los Angeles','San Diego','Portland','Washington','San Francisco','San Jose','Austin','Las Vegas']

#2 Scrap data from https://www.redfin.com/
for loc in Cities:
    property_urls2=[]
    driver=webdriver.Chrome()
    driver.get("https://www.redfin.com/")
    location=driver.find_element(By.XPATH,'//*[@id="search-box-input"]')
    location.send_keys(loc)
    search=driver.find_element(By.XPATH,'//div[@class="search-container inline-block"]/button')
    search.click()
    time.sleep(3)
    for _ in range(50):
        driver.execute_script("window.scrollBy(0,20)")
        time.sleep(3)
    start=0
    end=1
    for page in range(start,end):
        url2=driver.find_elements(By.XPATH,'//div[@class="bp-Homecard bp-InteractiveHomecard MapHomecardWrapper MapHomecardWrapper--photosLoading bp-InteractiveHomecard--hideNumIndicator clickable"]/a')
        for i in url2:
            property_urls2.append(i.get_attribute("href"))
        time.sleep(3)
    print(property_urls2)
    for url2 in property_urls2:
        driver=webdriver.Chrome()
        driver.get(url2)
        time.sleep(3)
        
        try:
            bhk=driver.find_element(By.XPATH,'//*[@id="content"]/div[12]/div[2]/div[1]/div/div/div/div[1]/div[2]/div/div/div/div/div[2]/div')
            BHK.append(''.join(i for i in bhk.text if i.isdigit()))
        except NoSuchElementException:
            BHK.append(np.nan)
            
        try:
            property=driver.find_element(By.XPATH,'//*[@id="content"]/div[12]/div[2]/div[9]/div/div/div/div/div/div[2]/div')
            Property.append(property.text)
        except NoSuchElementException:
            Property.append(np.nan)
            
        try:
            beds=driver.find_element(By.XPATH,'//*[@id="content"]/div[12]/div[2]/div[1]/div/div/div/div[1]/div[2]/div/div/div/div/div[2]/div')
            Beds.append(''.join(i for i in beds.text if i.isdigit()))
        except NoSuchElementException:
            Beds.append(np.nan)
            
        try:
            baths=driver.find_element(By.XPATH,'//*[@id="content"]/div[12]/div[2]/div[1]/div/div/div/div[1]/div[2]/div/div/div/div/div[3]/div')
            Baths.append(''.join(i for i in baths.text if i.isdigit()))
            Balcony.append('2')
        except NoSuchElementException:
            Baths.append(np.nan)
            Balcony.append('1')
        
        try:
            address=driver.find_element(By.XPATH,'//*[@id="content"]/div[12]/div[2]/div[1]/div/div/div/div[1]/div[2]/div/div/div/header/div/h1/div[1]')
            Address.append(address.text)
        except NoSuchElementException:
            Address.append(np.nan)
        
        try:
            city=driver.find_element(By.XPATH,'//*[@id="content"]/div[12]/div[2]/div[1]/div/div/div/div[1]/div[2]/div/div/div/header/div/h1/div[2]')
            pattern1=","
            pattern2=" "
            result1=re.split(pattern1,city.text)
            City.append(result1[0])
            result2=re.split(pattern2,result1[1])
            State.append(result2[1])
            Country.append('USA')
        except NoSuchElementException:
            City.append(np.nan)
            State.append(np.nan)
            Country.append('USA')
    
        try:
            landmarks=driver.find_element(By.XPATH,'//*[@id="content"]/div[12]/div[2]/div[9]/div/div/div/div/div/div[7]/div')
            Landmarks.append(landmarks.text)
        except NoSuchElementException:
            Landmarks.append(np.nan)
        
        try:
            carpet=driver.find_element(By.XPATH,'//*[@id="content"]/div[12]/div[2]/div[1]/div/div/div/div[1]/div[2]/div/div/div/div/div[4]')
            Carpet.append(''.join(i for i in carpet.text if i.isdigit()))
        except NoSuchElementException:
            Carpet.append(np.nan)
            
        try:
            status=driver.find_element(By.XPATH,'//*[@id="content"]/div[12]/div[2]/div[1]/div/div/div/div[1]/div[1]/span')
            Status.append(status.text)
        except NoSuchElementException:
            Status.append(np.nan)
            
        try:
            transaction=driver.find_element(By.XPATH,'//*[@id="content"]/div[12]/div[2]/div[1]/div/div/div/div[1]/div[1]')
            Transaction.append(transaction.text)
        except NoSuchElementException:
            Transaction.append(np.nan)
        
        try:
            year=driver.find_element(By.XPATH,'//*[@id="content"]/div[12]/div[2]/div[9]/div/div/div/div/div/div[3]/div')
            Year.append(''.join(i for i in year.text if i.isdigit()))
        except NoSuchElementException:
            Year.append(np.nan)
    
        try:
            price=driver.find_element(By.XPATH,'//*[@id="content"]/div[12]/div[2]/div[9]/div/div/div/div/div/div[4]/div')
            Price.append(''.join(i for i in price.text if i.isdigit()))
        except NoSuchElementException:
            Price.append(np.nan)
        
        Ownership.append(np.nan)      
       
        try:
            parking=driver.find_element(By.XPATH,'//*[@id="content"]/div[12]/div[2]/div[9]/div/div/div/div/div/div[5]/div')
            Parking.append(parking.text)
        except NoSuchElementException:
            Parking.append(np.nan)

        try:
            booking=driver.find_element(By.XPATH,'//*[@id="MortgageCalculator"]/div/div[2]/div[2]/div/div/div/div[1]/div/div/div[2]/div/div/div/div[2]/div[2]/div[2]/div[2]')
            Booking.append(''.join(i for i in booking.text if i.isdigit()))
        except NoSuchElementException:
            Booking.append(np.nan)
            
        try:
            buy=driver.find_element(By.XPATH,'//*[@id="content"]/div[12]/div[2]/div[1]/div/div/div/div[1]/div[2]/div/div/div/div/div[1]/div')
            Buy.append(''.join(i for i in buy.text if i.isdigit()))
        except NoSuchElementException:
            Buy.append(np.nan)
```

```python
df1=pd.DataFrame({'BHK':BHK,'Property':Property,'Beds':Beds,'Baths':Baths,'Balcony':Balcony,'Address':Address,'City':City,'State':State,
                'Country':Country,'Landmarks':Landmarks})
df2=pd.DataFrame({'Carpet':Carpet,'Status':Status,'Floor':Floor,'Transaction':Transaction,'Year':Year,'Price':Price})
df3=pd.DataFrame({'Furnished':Furnished,'Facing':Facing,'Parking':Parking,'Ownership':Ownership,'Booking':Booking,'Buy':Buy})
df1.insert(0, 'Sno', range(1, 1 + len(df1)))
df2.insert(0, 'Sno', range(1, 1 + len(df2)))
df3.insert(0, 'Sno', range(1, 1 + len(df3)))
df1.set_index('Sno',inplace=True)
df2.set_index('Sno',inplace=True)
df3.set_index('Sno',inplace=True)
df1.to_csv('Table1.csv')
df2.to_csv('Table2.csv')
df3.to_csv('Table3.csv')
```
